```
data_df = base_Df4.withColumn("Number_of_villages(Inhabited)", col("Number_of_villages(Inhabited)").cast("long")).withColumn("Number_of_villages(Uninhabited)", col("Number_of_villages(Uninhabited)").cast('long')).withColumn("Number_of_towns", col("Number_of_towns").cast('long')).withColumn("Number_of_households", col("Number_of_households").cast('long')).withColumn("Population_(Persons)", col("Population_(Persons)").cast('long')).withColumn("Population_(Males)", col("Population_(Males)").cast('long')).withColumn("Population_(Females)", col("Population_(Females)").cast('long')).withColumn("Area_(In_sq_km)", col("Area_(In_sq_km)").cast('long')).withColumn("Population_per_sq_km", col("Population_per_sq_km").cast('long'))
display(data_df)
```
