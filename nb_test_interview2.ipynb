- Find the total number of groups in the state - DADRA & NAGAR HAVELI

```
DNH_group_df = data_df.filter(col("Area_Name")=="STATE - DADRA & NAGAR HAVELI ").select(col("Uni_code"),col("Group"),col("Nic_Name")).distinct()
display(DNH_group_df)

print("DADRA & NAGAR HAVELI has " +  str(DNH_group_df.count()) + " groups")
```

- Find the total number of MALE workers working in the "Mixed farming" industry in any 5 different states of India.

```
from pyspark.sql import functions as F
mixed_farming_males_df = data_df.filter((col("Area_Name")!="INDIA") & (col("Nic_Name")=="Mixed farming") & (col("Total/Rural/Urban")=="Total")).groupBy("Area_Name").agg(F.sum(data_df.TW_Main_Males+data_df.TW_Marginal_Males )).withColumnRenamed("sum((TW_Main_Males + TW_Marginal_Males))","Total_Male_Workers").limit(5)

display(mixed_farming_males_df)
```

- Find the total number of FEMALE employees working in Rural area in 10 different states in the "Forestry and logging" and "Fishing" industries.

```
Foresty_Logging_Fishing_df = data_df.filter((col("Area_Name")!="INDIA") & ((col("Nic_Name")=="Forestry and logging") | (col("Nic_Name")=="Fishing")) & (col("Total/Rural/Urban")=="Rural")).groupBy("Area_Name").agg(F.sum(data_df.TW_Main_Females+data_df.TW_Marginal_Females )).withColumnRenamed("sum((TW_Main_Females + TW_Marginal_Females))","Total_Female_Workers").limit(10)

display(Foresty_Logging_Fishing_df)
```

- Find names of all industries and states where the count of single workers in both "Main Workers" and "Marginal Workers" are more than 50,000.

```
single_workers_df = data_df.filter((col("Area_Name")!="INDIA") & ( col("Nic_Name")!="Total") &(col("Total/Rural/Urban")=="Total")).groupBy("Area_Name","Nic_Name").agg(F.sum(data_df.SW_Main_Persons+data_df.SW_Marginal_Persons )).withColumnRenamed("sum((SW_Main_Persons + SW_Marginal_Persons))","Total Single Worker").filter(col("Total Single Worker")> 50000)

# joined_single_worker_df = single_workers_df.join(data_df,single_workers_df.Area_Name ==  data_df.Area_Name,"inner").filter(col("Total Single Worker")>5000)
display(single_workers_df)
# display(joined_single_worker_df.dropDuplicates(["Area_Name","Nic_Name"]))
```
